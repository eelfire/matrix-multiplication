```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('results.csv')
```

```python
plt.figure(figsize=(10, 5), dpi=200)
sns.set(style="darkgrid")

ax = sns.barplot(x="Method", y="Execution Time (s)", hue="Method", data=df)
ax.set(xlabel="Benchmark", ylabel="Time (s)")
ax.set_xticklabels(ax.get_xticklabels(), rotation=25, horizontalalignment="right")
plt.tight_layout()

# Print values on top of bars
for p in ax.patches:
    ax.annotate(
        format(p.get_height(), ".2f"),
        (p.get_x() + p.get_width() / 2.0, p.get_height()),
        ha="center",
        va="center",
        xytext=(0, 10),
        textcoords="offset points",
    )

plt.savefig("results.png")
plt.show()
```
